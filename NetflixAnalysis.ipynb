```python
# Install necessary libraries
# You can run these commands in your terminal or command prompt

!pip install pandas scikit-learn

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_extraction.text import CountVectorizer

# Load the dataset
file_path = '/Users/gavinkok/Documents/AI ML Course/NetflixAnalysis/netflix_titles.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')
```

```python
# Example preprocessing (assuming 'description' and 'type' columns)
df = df.dropna(subset=['description', 'type'])
X = df['description']
y = df['type']
```

```python
# Convert text data to numerical data
vectorizer = CountVectorizer(stop_words='english')
X_vectorized = vectorizer.fit_transform(X)
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.3, random_state=42)
```

```python
# Train the Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
#Objective: Suggest movies or TV shows to users based on the content of the titles and descriptions.

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Compute TF-IDF vectors
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['description'])

# Compute similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Example: Get recommendations for a specific title
def get_recommendations(title):
    idx = df.index[df['title'] == title].tolist()[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Get top 10 similar titles
    movie_indices = [i[0] for i in sim_scores]
    return df['title'].iloc[movie_indices]

print(get_recommendations('Stranger Things'))
```

```python
#Objective: Analyze the sentiment of the descriptions to understand the tone of the content.

from textblob import TextBlob

# Example sentiment analysis
df['sentiment'] = df['description'].apply(lambda x: TextBlob(x).sentiment.polarity)
print(df[['title', 'sentiment']].head())
```

```python
#Objective: Group similar titles based on their descriptions or other features to discover content segments.
    
from sklearn.cluster import KMeans

# Compute TF-IDF vectors
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['description'])

# Apply K-Means Clustering
num_clusters = 5
km = KMeans(n_clusters=num_clusters, random_state=42)
km.fit(tfidf_matrix)
df['cluster'] = km.labels_

# Example: Show titles in each cluster
for i in range(num_clusters):
    print(f"Cluster {i}:")
    print(df[df['cluster'] == i]['title'].head(10))
```

```python
# Objective: Analyze how the number of titles or their characteristics (e.g., genres, ratings) have changed over time.
import matplotlib.pyplot as plt

# Trend Analysis
release_year_counts = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.plot(release_year_counts.index, release_year_counts.values, marker='o')
plt.title('Number of Titles Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
# Objective: Determine the popularity and trends of different genres over time.


import seaborn as sns

# Example: Genre popularity over years
df['genres'] = df['listed_in'].str.split(', ')
df_exploded = df.explode('genres')
genre_trends = df_exploded.groupby(['release_year', 'genres']).size().unstack().fillna(0)

plt.figure(figsize=(14, 8))
sns.lineplot(data=genre_trends)
plt.title('Genre Popularity Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
